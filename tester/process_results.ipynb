```python
import os
import json
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
RESULTS_DIR = "../star_home_dir/kucerj56"

FILES = "graph-hard/graf-24-23.txt", "graph-hard/graf-25-16.txt", "graph-hard/graf-23-20.txt"
```

```python
results: list[dict] = []

for file in os.listdir(RESULTS_DIR):
    file_path = os.path.join(RESULTS_DIR, file)
    if 500 < os.path.getsize(file_path) < 800 and os.path.isfile(file_path):
        print(file_path)
        with open(file_path, "r") as f:
            file_contents_str = f.read()
            file_contents = json.loads(file_contents_str)
            results.append(file_contents)
            print(file_contents)
```

```python
# SEQUENTIAL, DATA, TASK, OPEN_MPI
sequential_results = list(sorted(filter(lambda x: x["VERSION"] == "SEQUENTIAL", results), key=lambda x: -x["RUNTIME"]))
task_results = list(sorted(filter(lambda x: x["VERSION"] == "TASK", results), key=lambda x: -x["RUNTIME"]))
data_results = list(sorted(filter(lambda x: x["VERSION"] == "DATA", results), key=lambda x: -x["RUNTIME"]))
open_mpi_results = list(sorted(filter(lambda x: (x["VERSION"] == "OPEN_MPI" and x['THREAD_COUNT'] == 12), results), key=lambda x: x["PROCESS_COUNT"]))
```

```python
table_data = []

for f in FILES:
    # graph_data_results = list(filter(lambda x: f in x["FILE"], data_results))
    graph_data_sequential = list(filter(lambda x: f in x["FILE"], sequential_results))
    graph_data_task_results = list(filter(lambda x: f in x["FILE"], task_results))
    graph_data_data_results = list(filter(lambda x: f in x["FILE"], data_results))
    graph_data_open_mpi_results = list(filter(lambda x: f in x["FILE"], open_mpi_results))
    graph_data_results = list(filter(lambda x: f in x["FILE"], open_mpi_results))
    print(graph_data_results)
    # labels = [r["THREAD_COUNT"] for r in graph_data_results]
    labels = [f"P{r['PROCESS_COUNT']}T{r['THREAD_COUNT']}" for r in graph_data_results]
    data = [r["RUNTIME"] for r in graph_data_results]

    assert len(graph_data_sequential) == 1

    # df = pd.DataFrame(data={'Name': ['variant', 'thread_count', 'process_count', 'runtime', 'zrychleni', 'efektivita']})
    df = pd.DataFrame(columns=['variant', 'thread_count', 'process_count', 'runtime', 'zrychleni', 'efektivita'])
    i = 0
    s = graph_data_sequential[0]
    df.loc[i] = (
        {
            'variant': s["VERSION"],
            'thread_count': None,
            'process_count': None,
            'runtime': s['RUNTIME'],
            'zrychleni': None,
            'efektivita': None,
         }
    )
    i+=1

    for r in graph_data_task_results:
        df.loc[i] = (
        {
            'variant': r["VERSION"],
            'thread_count': r["THREAD_COUNT"],
            'process_count': None,
            'runtime': r['RUNTIME'],
            'zrychleni': s['RUNTIME'] / (r['RUNTIME']),
            'efektivita': s['RUNTIME'] / (r['RUNTIME'] * r['THREAD_COUNT']),
         }
        )
        i+=1

    for r in graph_data_data_results:
        df.loc[i] = (
        {
            'variant': r["VERSION"],
            'thread_count': r["THREAD_COUNT"],
            'process_count': None,
            'runtime': r['RUNTIME'],
            'zrychleni': s['RUNTIME'] / (r['RUNTIME']),
            'efektivita': s['RUNTIME'] / (r['RUNTIME'] * r['THREAD_COUNT']),
         }
        )
        i+=1

    for r in graph_data_open_mpi_results:
        df.loc[i] = (
        {
            'variant': r["VERSION"],
            'thread_count': r["THREAD_COUNT"],
            'process_count': r["PROCESS_COUNT"],
            'runtime': r['RUNTIME'],
            'zrychleni': s['RUNTIME'] / (r['RUNTIME']),
            'efektivita': s['RUNTIME'] / (r['RUNTIME'] * r['THREAD_COUNT'] * r["PROCESS_COUNT"]),
         }
        )
        i+=1


    print(df)

    cur_table = ""

    for _, r in df.iterrows():
        # print(r)
        line = [
            str(r['variant']),
            str(r['process_count']) if r['process_count'] is not None else "",
            str(r['thread_count']) if r['thread_count'] is not None else "",
            str(round(r['runtime'], 2)),
            str(round(r['zrychleni'], 2)) if r['zrychleni'] is not None else "",
            str(round(r['efektivita'], 2)) if r['efektivita'] is not None else "",
        ]
        cur_table += " & ".join(line) + "    \\\\ \n"
        print(cur_table)
    table_data.append(cur_table)


    for r in graph_data_results:
        print(f"{f} took {r['RUNTIME']}")


    # objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
    y_pos = np.arange(len(labels))
    # performance = [10,8,6,4,2,1]

    plt.bar(y_pos, data, align='center', alpha=0.5)
    plt.xticks(y_pos, labels)
    plt.ylabel('Runtime (sec.)')
    plt.title(f)

    plt.show()
```
